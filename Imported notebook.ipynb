# Coronavirus progression in Mexico

We import the libraries and data. We are importing data from an excel given by ecdc. 

```python
import pandas as pd
import seaborn as sns

url = 'https://www.ecdc.europa.eu/sites/default/files/documents/COVID-19-geographic-disbtribution-worldwide-2020-03-26.xlsx'
df_raw = pd.read_excel(url, engine='openpyxl')

df_raw['Countries and territories'].unique()
```

Selection of Mexico´s data to analyze. And processing data.

```python
df = df_raw

df = df[df['Countries and territories']=='Mexico']

df = df.sort_values(['Year', 'Month' ,'Day'], ascending=[1, 1, 1])

df['ts'] = pd.to_datetime(df[['Year', 'Month' ,'Day']])

```

Plotting of data. Also, we modified the dimensions of the plot to be visually friendly.

```python
sns.lineplot(x='ts', y='Cases', data=df, hue='Countries and territories')
sns.set(rc = {'figure.figsize':(15,5)})
```
